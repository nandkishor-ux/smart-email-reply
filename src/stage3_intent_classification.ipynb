```python
import os
import pandas as pd
from pathlib import Path

print(" Imports done!")
```

```python
BASE_DIR   = Path(os.getcwd()).parent
INPUT_CSV  = BASE_DIR / "data" / "processed" / "emails_preprocessed.csv"
OUTPUT_CSV = BASE_DIR / "data" / "processed" / "emails_with_intent.csv"

print(" Input :", INPUT_CSV)
print(" Output:", OUTPUT_CSV)
```

```python
df = pd.read_csv(INPUT_CSV)
print(f" Loaded {len(df):,} emails")
df.head(3)
```

```python
intent_keywords = {
    "meeting_request": [
        "meet", "meeting", "schedule", "call", "discuss",
        "appointment", "conference", "available", "calendar", "zoom",
        "catch up", "sync", "standup", "touchbase", "invite",
        "slot", "time", "reschedule", "postpone", "agenda"
    ],
    "complaint": [
        "issue", "problem", "error", "wrong", "complaint",
        "urgent", "fix", "broken", "failed", "mistake",
        "bug", "crash", "not working", "delay", "disappointed",
        "unacceptable", "terrible", "worst", "frustrating", "refund"
    ],
    "appreciation": [
        "thank", "thanks", "appreciate", "grateful",
        "excellent", "great job", "well done", "good work",
        "amazing", "fantastic", "outstanding", "brilliant",
        "kudos", "impressive", "wonderful", "superb", "proud"
    ],
    "apology": [
        "sorry", "apology", "apologize", "regret",
        "forgive", "excuse", "pardon", "fault",
        "inconvenience", "mistake", "oversight", "correction",
        "unfortunately", "missed", "delayed", "failed to"
    ],
    "information_request": [
        "please", "could", "would", "request", "need",
        "send", "provide", "share", "update", "status",
        "let me know", "inform", "clarify", "details",
        "question", "query", "asking", "wanted to know", "confirm"
    ],
    "greeting": [
        "hello", "hi", "hey", "good morning", "good afternoon",
        "good evening", "hope you", "trust you", "greetings",
        "dear", "how are you", "hope this finds"
    ],
    "farewell": [
        "goodbye", "bye", "take care", "see you", "talk soon",
        "best regards", "warm regards", "sincerely", "cheers",
        "looking forward", "until next", "have a good"
    ],
}

print("Intent keywords defined!")
```

```python
def detect_intent(text):
    if not isinstance(text, str):
        return "general"
    text = text.lower()
    for intent, keywords in intent_keywords.items():
        for keyword in keywords:
            if keyword in text:
                return intent
    return "general"

print(" Function ready!")
```

```python
df['intent'] = df['body_processed'].apply(detect_intent)
print(" Done!")
print("\n Intent Distribution:")
print(df['intent'].value_counts())
```

```python
df.to_csv(OUTPUT_CSV, index=False)
print(f"Saved → {OUTPUT_CSV}")
print(f"Total rows: {len(df):,}")
```
